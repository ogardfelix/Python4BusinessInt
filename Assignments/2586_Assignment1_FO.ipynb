## Task 1 Instructions


````python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
````

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.


***Task a)***

````python
filtered_transactions = [item for item in transactions if item["price"] > 0 and item["quantity"] >= 0] #Task a)
````

````python
print(filtered_transactions) #sanity check
````

***Task b)***

````python
for item in filtered_transactions: #for each input in "filtered_transactions"
    item["total"]=item["price"]*item["quantity"] #creating the "total" value, defining it and applying it to each input

print(filtered_transactions) #sanity check
````

***Task c)***

````python
regional_revenue = {} #creating an empty dictionary 
for item in filtered_transactions:
    region = item["region"] #storing all the regions
    regional_revenue[region] = regional_revenue.get(region, 0) + item["total"] #this line goes through each input

#the 'get' command retrieves the current value of the region 'i' and adds the transactions total value to the existing region revenue. 
#In case of it being the first transaction for a region, it states '0' as the region value, '(region,0)'.


print(regional_revenue)
````

***Task d)***

````python
best_region = max(regional_revenue, key=regional_revenue.get)
best_value = regional_revenue[best_region]
print(f"The highest regional revenue is {best_value} in {best_region}") 
````

### Task 2: Creating a function (1p) **INSTRUCTIONS**

You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
````

````python
def sales_summary(sales):
    total = sum(sales)
    average = sum(sales)/len(sales)
    best_day = sales.index(max(sales))+1 #+1 as the indexing starts from 1
    worst_day = sales.index(min(sales))+1
    above_average_count = sum(1 for i in sales if i > average) #counts the sum of days above the average. Each day that meets the condition gets a value of 1.
    

    return {
    "Total":total,
    "Average":average,
    "Best_Day":best_day,
    "Worst_Day":worst_day,
    "Above_Average_Count":above_average_count
    }

result = sales_summary(sales)
print(result)
````

### Task 4: Class and methods (1p) **INSTRUCTIONS**

Design a class called `Portfolio` that represents an investment portfolio. The constructor is provided for you:

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
```

Each key in `holdings` is a ticker symbol (string) and each value is a dictionary with keys `"shares"` and `"price_per_share"`. Add the following methods to the class:

a) A method `add_asset(ticker, shares, price_per_share)` that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method `total_value()` that returns the total market value of the portfolio.

c) A method `most_valuable()` that returns the ticker of the asset with the highest total value (shares × price).

d) A `__str__` method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

````python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    #Task a)
        
    def add_asset(self, ticker, shares, price_per_share):
        if ticker in self.holdings: 
            self.holdings[ticker]["shares"]+=shares #Updates the number of shares
            self.holdings[price_per_share]["price_per_share"]=price_per_share #updates the price of the asset
        else: #if the asset doesn't exist the code below adds it to 'holdings' with the given inputs
            self.holdings[ticker] = {
                "shares": shares,
                "price_per_share": price_per_share
            }

    #Task b)
    #for each ticker it calculates the total value, and adds the ticker's value to the running total of all assets in the portfolio 
    def total_value(self):
        total = 0
        for ticker in self.holdings: 
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            total += shares * price
        return total

    #Task c)

    #loops through each asset 'for ticker in self.holdings' ,calculates the asset value and compares it to the 'best_value' at the time,
    #until it finds the asset with the highest value.
    
    def most_valuable(self):
        best_ticker = None
        best_value = 0

        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            value = shares * price

            if value > best_value:
                best_value = value
                best_ticker = ticker

        return best_ticker

    #Task d)
    
    def __str__(self):
        return f"Owner: {self.owner}, Number of assets: {len(self.holdings)}, Total value: {self.total_value()}"

            
        
````

````python
tickers=["AAHD", "APPL", "TSLA"] #a list of my random tickers
prices_per_share=[50,40,30] #random share prices
shares=[20,100,55] #random amounts

#adding the assets to my portfolio using list indexes

portfolio=Portfolio("Felix")
portfolio.add_asset(tickers[0], shares[0], prices_per_share[0])
portfolio.add_asset(tickers[1], shares[1], prices_per_share[1])
portfolio.add_asset(tickers[2], shares[2], prices_per_share[2])


print(portfolio)
print(portfolio.most_valuable())
print(portfolio.total_value())
````

### Task 8: More Pandas (1p)

A company records daily website traffic over four weeks. Generate the data by running the following script first:

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

Using this generated data, perform the following:

a) Create a DataFrame with columns: `date`, `visitors`, `conversion_rate`, and `avg_order_value`. Set `date` as the index.

b) Compute two new columns: `conversions` (= visitors × conversion_rate, rounded to the nearest integer) and `daily_revenue` (= conversions × avg_order_value).

````python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
````

***Task a)***

````python
df = pd.DataFrame({'date':days, 'visitors':visitors,'conversion_rate':conversion_rate,'avg_order_value':avg_order_value}) #creating the dataframe by using the defined columns
df.set_index("date",inplace=True) #setting the index
df.head() #sanity check
````

***Task b)***

````python
conversions = round(df["visitors"]*df["conversion_rate"]) #defining the conversions

df["conversions"]=conversions #adding it to the dataframe to be able to compute the next line

daily_revenue = df["conversions"]*df["avg_order_value"] #defining daily revenue

df["daily_revenue"]=daily_revenue #adding it to the df
df
````
